```python
plt.figure(figsize=(8,5))

ax = sns.barplot(final_trial, x= 'insuline_name',y='hba1c_change',hue='adverse_reaction')
for bar in ax.containers:
    ax.bar_label(bar, padding =10)

plt.legend(bbox_to_anchor = [1,0.15,0.15,1])

```

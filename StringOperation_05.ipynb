```python
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
titanic['Last_Name'] = titanic.Name.str.split(',').str.get(0)
titanic.head()
```

```python
titanic.Name.str.split(',').str.get(1).str.strip().str.split(' ', n=1)
```

```python
titanic[['Title', 'First_Name']]=titanic.Name.str.split(',').str.get(1)\
.str.strip().str.split(' ', n=1, expand=True)
```

```python
titanic.head()
```

# replace string

```python
titanic.Title.value_counts()
```

```python
titanic.Title = titanic.Title.str.replace('Ms.', 'Miss.')
```

# Finding Things

```python
titanic.Last_Name[titanic.Last_Name.str.startswith('A')].head()
```

```python
titanic.Last_Name[titanic.Last_Name.str.endswith('a')].head()
```

```python
# find() : return lowest index point
# -1 defines there is no substring
titanic.Last_Name.str.find('a', start=3, end=5)   # start=3, end=5
```

```python
# slice(start, stop, step) : create substring by range
titanic.First_Name.str.slice(2, 15, 2)
```

# Searching 

```python
titanic['Mr.' == titanic.Name]
```

```python
# .contains() : helps to findout content with patt


titanic[titanic.Name.str.contains('Mr.', regex=False, case=False)]
```

## apply ()
-It's used to apply same methods to all elements of series at once.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.describe()
```

```python
tips.info()
```

```python
## Define functions
def billcategory(amt):
    if amt < 13:
        return 'low'
    elif amt>= 13 and amt < 24 :
        return 'mid'
    else:
        return 'high'
```

```python
billcategory(11)
```

```python
tips['bill_category']= tips.total_bill.apply(billcategory)
tips
```

```python
# size category
tips['size'].apply(lambda x: 'group' if x >1 else 'single')
```

## Display no.  of customers along with bill_category.

```python
tips.bill_category.value_counts()
```

# Grouping and Aggregation
- Grouping : group data by some columns or parameters.
-Aggregation : provide aggregrate values like sum,mean, min,etc
-Methods:
 - grouply : grouping
 -agg : apply multiple aggregrate methods.
 - pivot : multi-dimensional view.

# .groupby

```python
tip_sale=tips.groupby('day',observed=True)[['total_bill','tip']].sum()\
    .sort_values(by='total_bill')\
    .reset_index()
tip_sale
```

```python
tip_sale.total_bill.plot(kind='pie')
```

```python
tips.groupby(['day','time'],observed=True)[['total_bill','tip']].mean()
```

```python
tips.groupby(['time','day'],observed=True)[['total_bill','tip']].mean()
```

```python
tips.groupby(['day','time'],observed=True)[['total_bill','tip']].agg(['sum','mean'])
```

```python
tips.groupby(['day','time'],observed=True).agg({'total_bill':'mean','tip':'sum'})
```

## QN for Practise
- Categories tips as gorgeous for more than 5, standard for less than 5
- Convert male and female into Men and Women
- Calculate tops percentage by their bill amount

```python
#  Categories tips as gorgeous for more than 5, standard for less than 5

def tipsCategory(amt):
    if amt >=5 :
        return 'gorgeous'
    elif amt < 5 :
        return 'standard'
```

```python
tips['tips_category']=tips.tip.apply(tipsCategory)
tips
```

```python
# Convert male and female into Men and Women

mapping_gender = {'Male':'Men','Female':'Women'}
tips['gender_new']=tips.sex.map(mapping_gender)
tips
```

```python
# Calculate tops percentage by their bill amount

tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100
display(tips.head())
```

## Pivot Table
- It provides values as multidimensional view
- Default value is mean.

```python
pd.pivot_table(tips, values= ['tip','total_bill'], index=['sex','smoker']\
             ,observed = True,columns=['day'],aggfunc={'total_bill':'mean','tip':'sum'}\
             ,margins=True,margins_name='Total'
             )
```

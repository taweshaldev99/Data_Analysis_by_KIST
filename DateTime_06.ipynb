### Date/Time Arithmetic
- Perform calculations like adding/subtracting time or computing difference.
- Add/Subtract time : use pd.Timedelta or direct arithmetic

```python
df['date_after_2days'] = df.Date + pd.Timedelta('2 days 3 hours 15 minutes')
df
```

Time differences : subtract datetime columns to get time deltas

```python
df['time_diff']=df.date_after_2days - df.Date
df
```

```python
df.time_diff.dt.days
```

Resampling and Aggregation
- handle time series data by resampling(eg: daily to monthly)
- common frequencies: 'D'(day), 'W'(week), 'ME'(month-end), 'YE'(year-end)
- aggregation methods : mean(), sum(), max() etc.

```python
date_df = pd.DataFrame({
    'date': pd.date_range('2023-01-01', periods=10, freq= 'D'),
    'value': range(10,101, 10)
})
date_df.set_index('date', inplace=True)
date_df
```

```python
# resample to monthly average
date_df.resample('W').mean()
```

# Handling time zones
Pandas supports time zone-aware datetime objects.
- assign a time zone : use tz_localize()

```python
df.Date = df.Date.dt.tz_localize('UTC+05:45')
```

```python
df
```

- convert time zone : use tz_convert('define timezone')

```python
df.Date.dt.tz_convert('America/New_York')
```

```python
df.Date.dt.tz_convert('Europe/Berlin')
```

# Handling Missing Dates
- fill or interprate missing dates in a time series.
- reindex to include all dates

```python
df1 = pd.DataFrame({
    'date': pd.to_datetime(['2023-01-01', '2023-01-03']),
    'value': [42,50]
})
df1
```

```python
all_dates = pd.date_range('2023-01-01', '2023-01-03', freq='D')
df1.set_index('date').reindex(all_dates)
```

# Show different formats

```python
# Sample DataFrame with various date string formats
df = pd.DataFrame({
    'date_str': [
        '2025-09-02',              # ISO format
        '09/02/2025',             # US format
        '02-Sep-2025',            # Day-Month-Year
        '2025-09-02 12:05:00',    # Date and time
        '09/02/2025 12:05 PM',    # US format with AM/PM
        '2025-09-02 12:05:00+05:45'  # With timezone
    ]
})
df
```

# convert to datetime

```python
df['date'] = pd.to_datetime(df.date_str, format= 'mixed', utc=True)
df
```

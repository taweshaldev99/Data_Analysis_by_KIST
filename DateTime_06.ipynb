# DateTime Operation
- It is very important operation for time series analysis

```python
import numpy as np
import pandas as pd
```

### Create or change type of DateTime
- .to_datetime()

```python
data = {
    'Date': ['2023-01-01 10:12:15',
             '2023-01-02 11:12:13', '2023-01-03 11:12:13', '2023-01-04 11:12:13'],
    'Sales': [40000,50000, 35000, 55000]
}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.Date = pd.to_datetime(df.Date)
```

```python
# date_range(start, end, freq=)
new_date = pd.date_range('2023-01-01 0000:00:00', '2023-12-31 0000:00:00', freq='ME')
```

```python
new_df = pd.DataFrame({'Entry_date' : new_date})
new_df
```

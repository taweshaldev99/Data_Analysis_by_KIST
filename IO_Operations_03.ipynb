# create by random production

```python
from datetime import datetime, timedelta
import random

#set random seed for reproducibility
np.random.seed(42)
random.seed(42)

#generate data
ids= range(1,101)
names= [f'Emplyee_{i}' for i in range(1,101)]  # simple name format
ages = np.random.randint(22,60,size=100)
salary = np.random.randint(30000,120000,size=100)
departments = random.choices(['HR','IT','Finanace','Marketing','Operations'],k=100)
join_dates = [datetime(2020,1,1)+timedelta(days=random.randint(0,1825)) for _ in range(100)]

#Create dataframe
df=pd.DataFrame ({
    'Ids': ids,
    'Names': names,
    'Ages' : ages,
    'Salary' : salary,
    'Departments': departments,
    'Join_dates' :join_dates,
})
df
```

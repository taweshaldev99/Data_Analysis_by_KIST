#Frequency Counts
- It summarize how often each category or value appears in a dataset for a single variable.
- It's foundation of EDA(Exploratory Data Analaysis) for categorical data.
- It helps to find patterns distributions or imbalance in data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

## Frequency counts for passenger class

```python
titanic['class'].value_counts(normalize=True)*100
```

```python
sns.catplot(titanic, x='class',kind='count')
```

## Cross- Tabulation
- It extends frequency counts by multiple categories

```python
pd.crosstab([titanic['class'],titanic['sex']],titanic['survived'],margins=True,normalize='columns')*100
```
